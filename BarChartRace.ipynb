```python
import pandas as pd
from util.data import read_data

countries, data = read_data(update=True)
regions_ru = pd.read_csv('data/regions_ru.csv', index_col='country')
```

```python
FROM_DATE = '2020-03-25'
EXCLUDE_COUNTRIES = ['RU-MOW', 'RU-MOS', 'RU-SPE']

chart_data = data[data['country'].isin(regions_ru.index)][['date', 'country', 'confirmed']].copy()
chart_data = chart_data[chart_data['date'] >= FROM_DATE].sort_values(by='date')
chart_data['date'] = chart_data['date'].dt.strftime('%d.%m')

chart_data = chart_data.pivot(index='country', columns='date')['confirmed'][chart_data['date'].unique()]
chart_data = regions_ru.join(chart_data).drop(EXCLUDE_COUNTRIES).fillna(0)

# export to https://flourish.studio/
chart_data.to_csv('chart_data.csv', index=False)

chart_data.head()
```
