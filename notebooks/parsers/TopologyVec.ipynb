```python
import numpy as np
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/ods-ai-ml4sg/covid19-tutu/master/data/Covid-19-tutu-Region-Vectors.csv'
region_vec = pd.read_csv(url, sep=';', index_col=0)

region_vec['Region'] = region_vec['Region'].replace({
    'Adygea': 'RU-AD',
    'Altai': 'RU-ALT',
    'Amur': 'RU-AMU',
    'Arkhangelsk': 'RU-ARK',
    'Astrakhan': 'RU-AST',
    'Bashkortostan': 'RU-BA',
    'Belgorod': 'RU-BEL',
    'Bryansk': 'RU-BRY',
    'Buryat': 'RU-BU',
    'Chechnya': 'RU-CE',
    'Chelyabinsk': 'RU-CHE',
    'Chukotka': 'RU-CHU',
    'Chuvash': 'RU-CU',
    'Crimea': 'UA-43',
    'Dagestan': 'RU-DA',
    'Gorno-Altay': 'RU-AL',
    'Irkutsk': 'RU-IRK',
    'Ivanovo': 'RU-IVA',
    'Kabardino-Balkar': 'RU-KB',
    'Kaliningrad': 'RU-KGD',
    'Kalmykia': 'RU-KL',
    'Kaluga': 'RU-KLU',
    'Kamchatka': 'RU-KAM',
    'Karelia': 'RU-KR',
    'Kemerovo': 'RU-KEM',
    'Khabarovsk': 'RU-KHA',
    'Khakassia': 'RU-KK',
    'Khanty-Mansi': 'RU-KHM',
    'Kirov': 'RU-KIR',
    'Komi': 'RU-KO',
    'Kostroma': 'RU-KOS',
    'Krasnodar': 'RU-KDA',
    'Krasnoyarsk': 'RU-KYA',
    'Kurgan': 'RU-KGN',
    'Kursk': 'RU-KRS',
    'Leningrad': 'RU-LEN',
    'Lipetsk': 'RU-LIP',
    'Magadan': 'RU-MAG',
    'Mari El': 'RU-ME',
    'Mordovia': 'RU-MO',
    'Moscow City': 'RU-MOW',
    'Moscow Oblast': 'RU-MOS',
    'Murmansk': 'RU-MUR',
    'Nenets': 'RU-NEN',
    'Nizhny Novgorod': 'RU-NIZ',
    'North Ossetia-Alania': 'RU-SE',
    'Novgorod': 'RU-NGR',
    'Novosibirsk': 'RU-NVS',
    'Omsk': 'RU-OMS',
    'Orenburg': 'RU-ORE',
    'Oryol': 'RU-ORL',
    'Penza': 'RU-PNZ',
    'Perm': 'RU-PER',
    'Primorsky': 'RU-PRI',
    'Pskov': 'RU-PSK',
    'Rostov': 'RU-ROS',
    'Ryazan': 'RU-RYA',
    'Saint Petersburg': 'RU-SPE',
    'Sakha': 'RU-SA',
    'Sakhalin': 'RU-SAK',
    'Samara': 'RU-SAM',
    'Saratov': 'RU-SAR',
    'Smolensk': 'RU-SMO',
    'Stavropol': 'RU-STA',
    'Sverdlovsk': 'RU-SVE',
    'Tambov': 'RU-TAM',
    'Tatarstan': 'RU-TA',
    'Tomsk': 'RU-TOM',
    'Tula': 'RU-TUL',
    'Tuva': 'RU-TY',
    'Tver': 'RU-TVE',
    'Tyumen': 'RU-TYU',
    'Udmurt': 'RU-UD',
    'Ulyanovsk': 'RU-ULY',
    'Vladimir': 'RU-VLA',
    'Volgograd': 'RU-VGG',
    'Vologda': 'RU-VLG',
    'Voronezh': 'RU-VOR',
    'Yamalo-Nenets': 'RU-YAN',
    'Yaroslavl': 'RU-YAR',
    'Yevrey': 'RU-YEV',
    'Zabaykalsky': 'RU-ZAB',
})

region_vec = region_vec.dropna()
region_vec = region_vec[region_vec['Region'].str.startswith('RU-') | region_vec['Region'].str.startswith('UA-')]
region_vec = region_vec.reset_index(drop=True).drop(columns=['County'])

def read_numpy(x, sep=' '):
    return np.fromstring(x.replace('\n','').replace('[','').replace(']','').replace('  ',' '), sep=sep)

region_vec['FeatureVectorGlobal'] = region_vec['FeatureVectorGlobal'].apply(read_numpy)
region_vec['RegionFeatureVectorGlobal'] = region_vec['RegionFeatureVectorGlobal'].apply(lambda x: read_numpy(x, sep=','))

region_vec.head()
```

```python
global_feature_vec = region_vec.groupby('Region')['FeatureVectorGlobal'].apply(lambda x: np.mean(x))
region_feature_vec = region_vec.groupby('Region')['RegionFeatureVectorGlobal'].apply(lambda x: np.mean(x))

global_feature_vec = pd.DataFrame(
    np.stack(global_feature_vec.values),
    index=global_feature_vec.index, 
    columns=[f'global_vec_{i}' for i in range(len(global_feature_vec[0]))]
)

region_feature_vec = pd.DataFrame(
    np.stack(region_feature_vec.values),
    index=region_feature_vec.index, 
    columns=[f'region_vec_{i}' for i in range(len(region_feature_vec[0]))]
)
```

```python
ru_regions_vec = global_feature_vec.join(region_feature_vec)
ru_regions_vec.to_csv('data/regions_ru_vec.csv')
ru_regions_vec.head()
```
